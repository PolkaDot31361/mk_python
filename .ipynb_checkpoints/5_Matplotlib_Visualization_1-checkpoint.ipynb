### Subplot 사용 <font color="red">(중요)</font>

```python
x = np.arange(0, 10, 0.01)

plt.subplot(2, 1, 1); plt.plot(x, np.sin(x)) # 2행 1열로 쪼개서 그릴때 첫 번째 sine그래프
plt.subplot(2, 2, 3); plt.plot(x, np.cos(x))
plt.subplot(224); plt.plot(x, np.cos(x)*np.sin(x))
plt.show()
```

```python
plt.subplots?
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0, 0].plot(x, np.sin(x))
axes[0, 1].plot(x, np.cos(x))
axes[1, 0].plot(x, np.tan(x))
axes[1, 1].plot(x, np.sin(x)**2 + np.cos(x)**2)
```

### Font 적용 <font color="red">(중요)</font>

```python
# font 적용 중요
font = {
    "family":"consolas"
    ,"color":"#63abff"
    ,"weight":"normal"
    ,'size':20
}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2*np.pi*x)*np.exp(-x)

plt.plot(x, y, '#6490e4')
plt.title('Damped Exponential Decay', fontdict = font)

plt.xlabel('time', fontdict = font)
plt.ylabel('voltage (mV)', fontdict = font)

plt.show()
```

```python
t = np.arange(0.01, 5.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.ylim(-1.5, 1.5)
plt.xlim(0.01, 5.0)
plt.show()
```

```python
# 중요
x = [1, 2, 3, 4]
y = [0, 5, 3, 4]
labels = ['Coffee', 'Is', 'Always', 'Right']

plt.plot(x,y)
plt.xticks(x, labels, rotation = 45) # 눈금 레이블 회전을 각도 또는 키워드로 지정
plt.margins(0.1)     # 눈금 레이블이 잘리지 않도록 간격 조정
plt.show()
```

### 범례 legend <font color="red">(중요)</font>

```python
# 범례 legend

a = b = np.arange(0, 3, 0.02)
c = np.exp(a)
d = c[::-1]   # c의 마지막 값을 제외하고 사용

fig, ax = plt.subplots()
ax.plot(a, c, c="#6490e4", ls='--', label='Coffee')
ax.plot(a, d, c="#63abff", ls=":", label='IS')
ax.plot(a, c+d, c="#103e78", label='GOOD')

legend = ax.legend(loc='upper center', shadow=True)

plt.show()
```

### 축 공유 <font color="red">(중요)</font>

### 이산형 데이터 표현 (bar chart)
> `bar`, `barh` 사용

```python
plt.bar?
```

```python
city = ['Los Angeles', 'New York', "Chicago", 'Texas', 'Miami']
y = [144, 49, 81, 64, 36]
x = np.arange(len(y))

plt.title('Coffee')
plt.bar(x, y, color="#6490e4")
plt.xticks(x, city, rotation = 45)

plt.xlabel('City')
plt.ylabel('Ice')

plt.show()
```

```python
city = ['Los Angeles', 'New York', "Chicago", 'Texas', 'Miami']
men_perc = [49, 48, 46, 63, 40]
women_perc = [100 - m for m in men_perc]

x = np.arange(len(city))
width = 0.35

fig, ax = plt.subplots()

prec_m = ax.bar(x - width/2, men_perc, width, label='MEN', color="#6490e4")
prec_f = ax.bar(x + width/2, women_perc, width, label="WOMEN", color="#103e78")

ax.set_ylabel("Percentage")
ax.set_title("Gender Percentage")
ax.set_xticks(x)
ax.set_xticklabels(city, rotation=45)
ax.legend()
plt.xlabel('CITY')

plt.show()
```

```python
city = ['Los Angeles', 'New York', "Chicago", 'Texas', 'Miami']
y = [144, 49, 81, 64, 36]
x = np.arange(len(y))

plt.title('Coffee')
plt.barh(x, y, color="skyblue")
plt.yticks(x, city, rotation = 45)

plt.xlabel('City')
plt.ylabel('Ice')

plt.show()
```

### Scatter Plot <font color="red">(중요)</font>
> 데이터의 분포를 파악하기 위하여 사용

```python
np.random.seed(1968081)

fig, ax = plt.subplots()
for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 50
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color
               ,alpha = 0.3, edgecolors='none')
    
ax.legend()
ax.grid(True)
plt.show()
```

### BoxPlot <font color="red">(중요)</font>

```python
np.random.seed(1968081)

spread = np.random.rand(50)*100
center = np.ones(25)*50
flier_high = np.random.rand(10)*100 + 100
flier_low = np.random.rand(10) - 100

data = np.concatenate((spread, center, flier_high, flier_low))

fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)

plt.show()
```
