### 1. 주어진 Seattle2014.csv 파일을 확인하고 메모리에 로드하시오

```python
sea = pd.read_csv("./files/data/Seattle2014.csv")
display(sea)
```

### 2. 강우량 많은날 상위 10일 출력

```python
sea.sort_values(by='PRCP', ascending=False).head(10)
```

### 3. 1월의 강수량만 출력하시오

```python
sea_jan = sea[sea['DATE'] < 20140200]
sea_jan
```

### 4. 1년의 강수량중 40 미만으로 내린날이 몇일인지 출력하시오

```python
count = len(sea.where(sea['PRCP'] < 40).value_counts())
count
```

### 5. 1월에 내린 평균 강수량

```python
sea_jan['PRCP'].mean()
```

### 6. 2월에 내린 평균 강수량
