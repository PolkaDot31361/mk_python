## 데이터 추가

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1, 5], 'B':[11, 12]}, index=['커피', '도넛'])
display(df1)

df2 = pd.DataFrame({'A':[11, 51], 'B':[111, 121]}, index=['피자', '라떼'])
display(df2)
```

```python
df3 = df1.append(df2)

display(df3)
```

```python
df4 = pd.DataFrame({'A':[45, 7], 'B':[13, 83]})
display(df4)
```

```python
df1.append(df4)
```

```python
# df5 = pd.DataFrame({'A':1, 'B':4})  error
df5 = pd.DataFrame({'A':[1], 'B':[4]})
display(df5)
```

```python
df6 = df1.append(df5)
display(df6)
```
