## [문제] 야구 게임
<pre>
- 컴퓨터가 0~9 사이의 중복 되지 않는 임의의 정수 3개를 생성
- 생성한 숫자와 순서까지 맞추는 게임 코드를 작성하시오
- 램덤하게 생성된 3개의 숫자, 플레어이가 입력한 숫자 3개에 대하여 순서와 숫자를 비교하여 다음의 결과를 보여준다.
* 숫자는 맞았으나 위치가 다른 경우, 맞춘 숫자 개수 만큼 '볼'로 표현
* 숫자와 순서까지 맞는 경우 ' 스트라이크'로 표현
* 게임은 3스트라이크가 되면 종료 되며, 시도 횟수를 보여준다.

- 0~9 사이의 중복되지 않는 정수 3개를 반환하는 randomNumber()함수를 정의 하여 사용
- .사용자로부터 0~9사이의 정수를 입력받아 반환 하는 inputNumber()함수를 정의하여 사용


******************Baseball Game*******************
각각의 입력 범위는 0~9입니다.
지금부터 게임을 시작합니다.
**************************************************
첫번째 값 : 1
두번째 값 : 2
세번째 값 : 3
count : 1 -> 현재 입력하신 내용에대한 결과는 n Strike, n Ball 입니다.
...
축하합니다 ! 총 시도 횟수는 n 번 입니다.
</pre>

```python
import random

def randomNumber():
    numbers = set()
    while True:
        if len(numbers)==3: break
        else:
            numbers.add(random.randint(0,9))
    return list(numbers)        

def inputNumber():      
    numbers = list()
    while True:
        try:
            if len(numbers) == 3: break
            elif len(numbers) == 0:
                first = int(input("첫번째 값 :"))
                numbers.append(first)
            elif len(numbers) == 1:
                second = int(input("두번째 값 : "))
                if second in numbers:
                    print("이미 입력한 값입니다.")
                else:
                    numbers.append(second)
            elif len(numbers) == 2:
                third = int(input("세번째 값 : "))
                if third in numbers:
                    print("이미 입력한 값입니다.")
                else:
                    numbers.append(third)
        except ValueError as e:
            print("정수를 입력해 주세요.")
    return list(numbers)

print("*"*15+"Baseball Game"+"*"*15)
print("각각의 입력 범위는 0~9 입니다.")
print("지금부터 게임을 시작합니다.")
print("*"*43)

cnt = 1
answer = randomNumber()
while True:
    input_list = inputNumber()
    strike = 0
    ball = 0
    
    if answer == input_list:
        print("축하합니다! 총 시도 횟수는 %d 번 입니다." %cnt)
        break
        
    for i in input_list:
        if i in answer:
            ball +=1
    for i in range(3):
        if answer[i] == input_list[i]:
            ball -= 1
            strike += 1
    print("count : %d -> 현재 입력하신 내용에대한 결과는 %d Strike, %d Ball 입니다." % (cnt, strike, ball))
    
    cnt += 1
    
```
