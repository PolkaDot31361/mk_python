# 1. Package Import

```python
import numpy as np
import pandas as pd
```

# 2. 두개의 파일을 읽어온 후, 파일의 정보 출력
> `place` : 대여소 정보\
> `rent_info` : 대여 정보

```python
place = pd.read_excel('./files/data/서울특별시 공공자전거 대여소 정보(19.12.9).xlsx')
rent_info = pd.read_csv('./files/data/서울특별시 공공자전거 대여정보_201911_2.csv')
```

```python
print(place.info())
print("="*45)
print(rent_info.info())
```

# 3. NaN, Null 값의 개수 확인
> NA값이 들어있는 데이터 삭제 `dropna`

```python
place.isna().sum()
```

# 4. Data Type 변환

```python
place['대여소ID'] = place['대여소ID'].astype('int32')
place.info()
```

```python
# place['기준시작일자'] = pd.to_datetime(place['기준시작일자'])
# '개통'이란 문자열이 들어가 있음, . 을 - 로 변경해 주어야 한다

place['기준시작일자'] = place['기준시작일자'].replace('개통', '', regex = True)
place['기준시작일자'] = pd.to_datetime(place['기준시작일자'])
place.info()
```

```python
place.head(10)
place.tail(10)
```

```python
rent_info
```

```python
rent_info['대여 대여소번호'] = rent_info['대여 대여소번호'].astype('int32')
rent_info['반납대여소번호'] = rent_info['반납대여소번호'].astype('int32')
rent_info['대여일시'] = pd.to_datetime(rent_info['대여일시'])
rent_info['반납일시'] = pd.to_datetime(rent_info['반납일시'])

rent_info.info()
```

# 6. 대여소 구별 거치대 수

```python
place.groupby(by="대여소_구")['거치대수'].sum()
```

# 7. 거치대 수의 합계, 평균, 최대, 최소 출력
>합계별 내림차순으로 정렬\
>10개 출력

```python
bi_grp = place.groupby(by='대여소_구')['거치대수'].agg([sum, np.mean, max, min])
```

```python
bi_grp.sort_values(by = 'sum', ascending = False).head(10)
```

```python
place.groupby(by='대여소_구')['거치대수'].agg([sum, np.mean, max, min]).sort_values(by = 'sum', ascending = False).head(10)
```

# 8. 대여소 정보 join 대여 정보
>`merge`사용\
>`left_on` = '대여소ID', `right_on` = '대여 대여소번호' 사용

```python
rent_bicycle = pd.merge(place, rent_info, left_on='대여소ID', right_on='대여 대여소번호')
# rent_bicycle = place.merge(rent_info, left_on='대여소ID', right_on='대여 대여소번호')
```

```python
display(rent_bicycle.head(10))
```

```python
rent_bicycle.info()
```

# 9. 각 구별 이용시간,  이동거리의 평균, 최대, 최소를 출력
> 구를 index로 사용

```python
rent_grp = rent_bicycle.groupby('대여소_구')
```

```python
rent_dis = rent_grp[['이용시간', '이용거리']].agg([np.mean, max, min]).round(2)
```

```python
rent_dis.sort_index().head(10)
```
